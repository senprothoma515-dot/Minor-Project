```python
import pandas as pd
df = pd.read_excel(r"C:\Users\senpr\OneDrive\Documents\classroom_data.xlsx")
print(df)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df.head()
df.tail()
print("Shape of dataset:", df.shape)

print("\nDataset Info:\n")
df.info()

df.isnull().sum()
df['Utilization_Percent'] = (df['Enrolled_Students'] / df['Capacity']) * 100
df.head()
underutilized_rooms = df[df['Utilization_Percent'] < 50]
underutilized_rooms
plt.figure(figsize=(10,5))
plt.bar(df['Room_ID'], df['Capacity'], label='Capacity')
plt.bar(df['Room_ID'], df['Enrolled_Students'], label='Enrolled Students')
plt.xlabel("Room ID")
plt.ylabel("Number of Students")
plt.title("Classroom Capacity vs Enrollment")
plt.legend()
plt.show()
df_small = df.head(10)

plt.figure(figsize=(10,5))
sns.barplot(x='Room_ID', y='Utilization_Percent', data=df_small)
plt.axhline(50, color='red', linestyle='--', label='50% Threshold')
plt.title("Classroom Utilization Percentage")
plt.legend()
plt.show()
labels = ['Utilized Rooms', 'Underutilized Rooms']
sizes = [
    len(df[df['Utilization_Percent'] >= 50]),
    len(df[df['Utilization_Percent'] < 50])
]

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Classroom Utilization Distribution")
plt.show()
df['Utilization_Percent'].describe()

print("Total Classrooms:", len(df))
print("Underutilized Classrooms:", len(underutilized_rooms))
print("Average Utilization:", df['Utilization_Percent'].mean())
```
